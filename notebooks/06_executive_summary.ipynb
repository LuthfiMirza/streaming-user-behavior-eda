```python
from pathlib import Path
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

DATA = Path('../data/processed/streaming_interactions_enriched.csv')
df = pd.read_csv(DATA, parse_dates=['started_at', 'signup_date', 'release_date'])
df.head()
```

## Starter Analysis
Use the cells below as a reproducible starting point for the chapter narrative.


```python
summary = df.groupby(['genre', 'format']).agg(
    total_playtime=('listened_minutes', 'sum'),
    listeners=('user_id', 'nunique'),
    avg_completion=('completion_rate', 'mean'),
    skip_rate=('skipped', 'mean'),
).reset_index().sort_values('total_playtime', ascending=False)
summary.head(10)
```

```python
plt.figure(figsize=(10, 5))
top_genres = df.groupby('genre')['listened_minutes'].sum().sort_values(ascending=False).head(8)
sns.barplot(x=top_genres.values, y=top_genres.index, color='#4C78A8')
plt.title('Top Genres by Total Playtime')
plt.xlabel('Listened Minutes')
plt.ylabel('Genre')
plt.tight_layout()
```
